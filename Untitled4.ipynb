```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
final = pd.DataFrame()

topic = ["travel_2" , "mystery_3" , "historical-fiction_4" , "sequential-art_5" , "classics_6" , "philosophy_7" , "romance_8" , "womens-fiction_9" , "fiction_10" , "childrens_11" , "religion_12" , "nonfiction_13" , "music_14" , "default_15" , "science-fiction_16" , "sports-and-games_17" , "add-a-comment_18" , "fantasy_19" , "new-adult_20"]

for i in topic:
    url = "https://books.toscrape.com/catalogue/category/books/{}/index.html".format(i)

    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36"
    }

    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, "lxml")

    total = soup.select("article.product_pod")

    names=[]
    rating=[]
    price=[]
    available=[]

    for i in total:

      names.append(i.select("h3 a")[0].text.strip())
      rating.append(i.find("p" , class_="star-rating").get("class")[1].strip())
      price.append(i.find_all("p" , class_="price_color")[0].text.strip())
      available.append(i.find_all("p" , class_="instock availability")[0].text.strip())


    d = {"Name": names, "Rating": rating, "Price": price, "Available": available}

    df = pd.DataFrame(d)

    final = pd.concat([final, df], ignore_index=True)


final.to_csv("books_dataset.csv", index=False)
```
